```python
target_cols = ["final_score", "grade", "pass_fail"]

academic_history_cols = [
    "previous_score",
    "math_prev_score",
    "science_prev_score",
    "language_prev_score"
]

behavior_cols = [
    "daily_study_hours",
    "attendance_percentage",
    "homework_completion_rate"
]

lifestyle_cols = [
    "sleep_hours",
    "screen_time_hours",
    "physical_activity_minutes"
]

psychological_cols = [
    "motivation_score",
    "exam_anxiety_score"
]

context_cols = [
    "parent_education_level",
    "study_environment"
]

numeric_cols = (
    academic_history_cols
    + behavior_cols
    + lifestyle_cols
    + psychological_cols
    + ["final_score"]
)
```

```python
df["student_readiness_score"] = (
    df["previous_score"] * 0.25 +
    df["daily_study_hours"] * 10 +
    df["attendance_percentage"] * 0.20 +
    df["homework_completion_rate"] * 0.20 +
    df["motivation_score"] * 5 -
    df["exam_anxiety_score"] * 5
)
df.tail()
```

```python
plt.figure()
sns.histplot(df["student_readiness_score"], kde=True)
plt.title("Student Readiness Score Distribution")
plt.show()
```

```python
df["performance_segment"] = pd.qcut(
    df["final_score"],
    q=4,
    labels=["Low","Mid","High","Top"]
)
```

```python
df.groupby("performance_segment")[
    [
        "daily_study_hours",
        "attendance_percentage",
        "sleep_hours",
        "screen_time_hours",
        "motivation_score",
        "exam_anxiety_score"
    ]
].mean()
```
